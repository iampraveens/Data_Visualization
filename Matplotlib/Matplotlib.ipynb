```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'Y Cubed')

ax.legend(loc=0)
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'Y Cubed')

ax.legend(loc=(0.1,0.2))
plt.show()
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
```

```python

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.set_title('title')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.plot(x,y)

```

```python

fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])

axes1.plot(x,y)
axes2.plot(x,y)


```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])

```

```python

fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])




```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (10,1.5))

axes[0].plot(x,y, color = 'blue', ls = '-', lw = 3)
axes[1].plot(z, color = 'red', ls = '--')
```
